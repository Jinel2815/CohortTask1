```python
import pandas as pd
import os
import numpy as np
import cv2
```

```python
df =pd.read_csv('data_labels.csv')
df
```

```python
folder=input()
```

```python
dir='output2'
wd=os.getcwd()
folder_path=os.path.join(wd,folder) 

if not os.path.isdir(dir):
    os.mkdir(dir)
    

for index , row in df.iterrows():
    

     
    
    image_name=str(row['name'])
    name='output' + image_name
    image_name=folder_path + '\\' + image_name
    img=cv2.imread(image_name)
    print(img)
    pt1=(row['xmin'],row['ymin'])
    pt2=(row['xmax'],row['ymax'])
    color =(0,255,0)
    thickness=5
    
    img1=cv2.rectangle(img, pt1, pt2, color, thickness)
    img1 = cv2.putText(img1, row['label'], (row['xmin'], row['ymin']-10), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,0,0), 2)
    
  
   
    if img1 is not None:
        cv2.imwrite(os.path.join(dir,name),img1)
    else:
        continue
        
        
        
    
    

    
    
```
